### Playlist id

```python
playlist_data=response
playlist_name = playlistName
playlist_id = None

for item in playlist_data['items']:
    if item['snippet']['title'] == playlist_name:
        playlist_id = item['id']
        break

if playlist_id:
    print(f"Playlist ID for '{playlist_name}': {playlist_id}")
else:
    print(f"Playlist '{playlist_name}' not found in the data.")
```

### Playlist items

```python
# Get the playlist items
playlist_items = []
next_page_token = None

while True:
    playlist_request = youtube.playlistItems().list(
        part='snippet',
        playlistId=playlist_id,
        maxResults=50,
        pageToken=next_page_token
    )
    playlist_response = playlist_request.execute()

    playlist_items.extend(playlist_response['items'])

    next_page_token = playlist_response.get('nextPageToken')

    if not next_page_token:
        break

# Extract video titles
video_titles = [item['snippet']['title'] for item in playlist_items]

# Print video titles
for title in video_titles:
    print(title)
```

### Video details

```python
# Set the video title you want to search for
video_title =video_title_user_input

# Search for the video by title
search_request = youtube.search().list(
    q=video_title,
    type='video',
    part='id',
    maxResults=1
)
search_response = search_request.execute()

# Check if there are any search results
if search_response['items']:
    video_id = search_response['items'][0]['id']['videoId']
    video_id_user_ip=video_id

    # Request video details by video ID
    video_request = youtube.videos().list(
        part='snippet,statistics',
        id=video_id
    )
    video_response = video_request.execute()

    # Check if there are video details
    if video_response['items']:
        video_details = video_response['items'][0]

        # Extract information
        video_title = video_details['snippet']['title']
        video_description = video_details['snippet']['description']
        video_view_count = video_details['statistics']['viewCount']
        video_like_count = video_details['statistics']['likeCount']

        print(f"Video Title: {video_title}")
        print(f"Video Description: {video_description}")
        print(f"View Count: {video_view_count}")
        print(f"Likes: {video_like_count}")
    else:
        print("Video details not found or unavailable.")
else:
    print("Video not found.")
```

### Top 10 most recent videos for the channel

```python
def get_channel_videos(channel_id, criteria, max_results=10):
    search_response = youtube.search().list(
        q=criteria,
        type='video',
        part='id',
        channelId=channel_id,
        maxResults=max_results,
        order='date'  # Change this to 'likeCount' or 'date' as needed
    ).execute()

    videos = []
    for search_result in search_response.get('items', []):
        video_id = search_result['id']['videoId']
        #video_title = search_result['snippet']['title']
        videos.append((video_id))

    return videos

# Set the channel ID for the specific YouTube channel you want to analyze
channel_id = channel_id_temp

# Get the top 10 most recent videos for the channel
most_recent = get_channel_videos(channel_id, criteria='', max_results=10)

print("\nTop 10 Most Recent Videos for the Channel:")
for video_id in most_recent:
    print(f": https://www.youtube.com/watch?v={video_id}")
```

### Perform sentiment analysis using textblob on the retrieved comments

#### example of using vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Sample text
text = "VADER is  a tool for sentiment analysis. I am using it!"

# Analyze sentiment
sentiment = analyzer.polarity_scores(text)
print(sentiment)

# Print the sentiment scores
print("Sentiment Scores:", sentiment)

# Determine sentiment label based on compound score
if sentiment['compound'] >= 0.05:
    sentiment_label = "Positive"
elif sentiment['compound'] <= -0.05:
    sentiment_label = "Negative"
else:
    sentiment_label = "Neutral"

print("Sentiment Label:", sentiment_label)
```
